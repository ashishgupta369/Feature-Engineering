```python
df['BsmtQual_NAN']=np.where(df['BsmtQual'].isnull(),1,0)
frequent=df['BsmtQual'].mode()[0]
df.loc[:,'BsmtQual'] = df['BsmtQual'].fillna(frequent)
```

```python
df['FireplaceQu_NAN']=np.where(df['FireplaceQu'].isnull(),1,0)
frequent=df['FireplaceQu'].mode()[0]
df.loc[:, 'FireplaceQu'] = df['FireplaceQu'].fillna(frequent)
```

```python
df['GarageType_NAN']=np.where(df['GarageType'].isnull(),1,0)
frequent=df['GarageType'].mode()[0]
df.loc[:,'GarageType']=df['GarageType'].fillna(frequent)
```

### suppose you have more frequent categories , we just replace NAN with new categories feature

```python
df=pd.read_csv('loan.csv',usecols=['BsmtQual','FireplaceQu','GarageType','SalePrice'])
```

```python
def imputate_nan(df,variable):
    df[variable+'_new']=np.where(df[variable].isnull(),'Missing',df[variable])
```

```python
for feature in ['BsmtQual','FireplaceQu','GarageType']:
    imputate_nan(df,feature)
```
